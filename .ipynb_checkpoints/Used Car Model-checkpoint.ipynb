```python
import pandas as pd
import numpy as np
import glob

import matplotlib.pyplot as plt
import seaborn as sns

from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.model_selection import train_test_split

from sklearn.metrics import r2_score
```

```python
url = 'https://raw.githubusercontent.com/AdamShafi92/Used-Car-Price/master/Dataset/'
```

```python
df_all = pd.DataFrame()

for f in glob.glob("/Users/User/Python Projects/Used Car Prices/Used-Car-Price/Dataset/*.csv"):

    df = pd.read_csv(f) # make sure to apply correct settings (sep, parse_dates, headers, missing_values)
    df["origin"] = f #add a column with a csv name
    df_all = df_all.append(df, ignore_index=True) #append new df to the "master" dataframe
```

```python
df_all['car'] = df_all['origin'].apply(lambda x: x.split("/")[7].replace('.csv',''))
df_all = df_all.drop(['origin'], axis =1)
```

```python
df_all['car'] =df_all['car'].replace('cclass','merc')
df_all['car'] =df_all['car'].replace('focus','ford')
```

```python
# model, transmission, fueltype and car are all string so we encode these
```

```python
data_encode_dummy = pd.get_dummies(df_all,columns=['model', 'transmission','fuelType','car'], drop_first = True)
data_encode_dummy
```

```python
# plot against price to check if the relationship is linear
```

```python
X = data_encode_dummy.drop(columns = ['price'])
Y = data_encode_dummy['price']

title = ['Price vs Year', 
         'Price vs Mileage', 
         'Price vs Tax', 
         'Price vs Mpg', 
         'Price vs Engine Size']

fig,ax = plt.subplots(3,2,figsize=(20,20))

i = 0
for rows in range(3):
    for cols in range(2):
        if rows == 2 and cols == 1:
            fig.delaxes(ax[rows,cols])
            break
        ax[rows,cols].scatter(x = X[X.columns[i]], y = Y)
        ax[rows,cols].set_title(title[i])
        i = i+1
        
fig.subplots_adjust(hspace=0.5, wspace=0.5)
```

```python
# the relationship is not linear and is closer to logarithmic so we can take the log of price
```

```python
data_encode_dummy_transform = data_encode_dummy.copy()
data_encode_dummy_transform['price'] = np.log(data_encode_dummy_transform['price'])
data_encode_dummy_transform
```

```python
X = data_encode_dummy_transform.drop(columns = ['price'])
Y = data_encode_dummy_transform['price']

title = ['Price vs Year', 
         'Price vs Mileage', 
         'Price vs Tax', 
         'Price vs Mpg', 
         'Price vs Engine Size']

fig,ax = plt.subplots(3,2,figsize=(20,20))

i = 0
for rows in range(3):
    for cols in range(2):
        if rows == 2 and cols == 1:
            fig.delaxes(ax[rows,cols])
            break
        ax[rows,cols].scatter(x = X[X.columns[i]], y = Y)
        ax[rows,cols].set_title(title[i])
        i = i+1
        
fig.subplots_adjust(hspace=0.5, wspace=0.5)
```

```python
# check multicolinearity - are variables related?
```

```python
X = data_encode_dummy_transform[['year','mileage','tax','mpg','engineSize']]
X.corr()
```

```python
# here we check Variance Inflation Factor - 
#this shows whether an independent variable can be predicted from another in the dataset
```

```python
VIF = pd.Series([variance_inflation_factor(X.values, i) for i in range(X.shape[1])], 
                index=X.columns)
print(VIF)
```

```python
X = data_encode_dummy_transform.drop(columns = ['price'])
Y = data_encode_dummy_transform['price']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)

X_train = X_train.reset_index(drop = True)
X_test = X_test.reset_index(drop = True)
Y_train = Y_train.reset_index(drop = True)
Y_test = Y_test.reset_index(drop = True)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor()
```

```python
model.fit(X_train,Y_train)
```

```python
model.score(X_train,Y_train)
```

```python
prediction = model.predict(X_test)
```

```python
data_test = {
    'Y_test' : Y_test,
    'Prediction' : prediction
}
```

```python
pd.DataFrame(data_test)
```

```python
mean_squared_error(Y_test, prediction)
```

```python
r2_score(Y_test, prediction)
```
